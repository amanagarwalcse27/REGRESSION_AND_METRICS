```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\AMAN\OneDrive\Desktop\python ml\Datasets\data.csv")
df.head()
```

```python
df.columns
```

```python
df.drop(columns= ['id','Unnamed: 32'],inplace= True)
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
print(x.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size= 0.2, random_state= 2)
X_train.head()
```

```python
#Preprocessing the data using scaling as KNN based on distance
from sklearn.preprocessing import StandardScaler
```

```python
print(X_train.dtypes)
```

```python
#creating object 
scaler = StandardScaler()

#fit the data using X_train
X_train = scaler.fit_transform(X_train)   #fit + transform on train data
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 3)
```

```python
#training testing the model
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
y_pred
```

```python
#Making matrix for this
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
#plotting graph
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Checking for different values for k
scores = []

for i in range(1,16):
    knn = KNeighborsClassifier(n_neighbors= i)

    knn.fit(X_train,y_train)

    y_pred = knn.predict(X_test)

    scores.append(accuracy_score(y_test,y_pred))
```

```python
plt.plot(range(1,16),scores)
plt.grid()
plt.xlabel("Value of K")
plt.ylabel("ACCURACY SCORES")
```
